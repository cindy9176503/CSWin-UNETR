# 2020 Marker-based Watershed Transform

```python
import numpy as np
from numpy.lib.function_base import append
import os

import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as patches

from PIL import Image, ImageDraw, ImageFont

import torch
import torch.utils.data
import torchvision.transforms as transforms
from torch.nn.functional import normalize

import SimpleITK as sitk

import imageio

import cv2 as cv

from scipy.ndimage import label, binary_dilation, binary_opening, binary_closing, binary_erosion, distance_transform_edt
from scipy import stats, ndimage

from skimage.segmentation import watershed, mark_boundaries, find_boundaries 
from skimage import io, morphology
import skimage.filters as filters
```

```python
pid = '1000915187_20191217'

model_name = 'cswinunetr_96_3_mask'
exp_name = 'exp_1_drop0_0_01_mask' 

root_exp_dir = os.path.join(workspace_dir, 'exps')

data_name = 'tooth'
mask_model = 'cswinunetr_96_3_mask'
mask_exp = 'exp_1_drop0_0_01_mask'

edge_model = 'cswinunetr_96_3_center_edge'
edge_exp = 'exp_1_drop0_0_01_edge'

cbct_path = f'/nfs/Workspace/dataset/teeth/data50/image/{pid}.nii.gz'.format(pid)
gt_path = f'/nfs/Workspace/dataset/teeth/data50/label/{pid}.nii.gz'.format(pid)
mask_path = f'{root_exp_dir}/{mask_model}/{data_name}/{mask_exp}/infers/{pid}.nii.gz'.format(pid)
edge_path = f'{root_exp_dir}/{edge_model}/{data_name}/{edge_exp}/infers/{pid}.nii.gz'.format(pid)

save_path = os.path.join(root_exp_dir, 'mwt_2020')
if not os.path.isdir(save_path):
    os.mkdir(save_path)
    

cbct_nii = sitk.ReadImage(cbct_path)
cbct_img = sitk.GetArrayFromImage(cbct_nii)

gt_nii = sitk.ReadImage(gt_path)
gt = sitk.GetArrayFromImage(gt_nii).astype(int)
gt_binary = gt.copy()
gt_labels_num = len(np.unique(gt))
gt_binary[gt>0] = 1
print('gt_labels_num = ', gt_labels_num)

pred_mask = sitk.ReadImage(mask_path)
pred_mask = sitk.GetArrayFromImage(pred_mask).astype(int)

pred_edge = sitk.ReadImage(edge_path)
pred_edge = sitk.GetArrayFromImage(pred_edge).astype(int)

sid = 130

plt.figure(dpi=200)
# plt.imshow(cbct_img[slice_id,:,:], cmap='gray')
# plt.axis('off')
# out_name = '{}/{}_cbct.jpg'.format(save_path, pid)
# plt.savefig(out_name)

plt.subplot(141)
plt.title('gt')
plt.axis('off')
plt.imshow(gt[sid,:,:])
plt.subplot(142)
plt.title('gt_binary')
plt.axis('off')
plt.imshow(gt_binary[sid,:,:], cmap='gray')
plt.subplot(143)
plt.title('pred_mask')
plt.axis('off')
plt.imshow(pred_mask[sid,:,:], cmap='gray')
plt.subplot(144)
plt.title('pred_edge')
plt.axis('off')
plt.imshow(pred_edge[sid,:,:], cmap='gray')
```

```python
# pred mask : Binarization
mask = pred_mask.copy()
mask[mask>=0.1] = 1
mask[mask<0.1] = 0
mask = mask.astype(np.uint8)

# pred mask : Erosion
pp_mask = mask.copy()
pp_mask = pp_mask.astype(np.uint8)
pp_mask = morphology.binary_erosion(pp_mask, morphology.ball(2))

# pred mask : Remove small objects
pp_mask = pp_mask.astype(bool)
pp_mask = morphology.remove_small_objects(pp_mask, min_size=500, connectivity=2)
pp_mask = pp_mask.astype(np.uint8)

plt.figure(dpi=200)
plt.rcParams['axes.titlesize'] = 5
plt.rcParams['axes.titlepad'] = 1
plt.subplot(121)
plt.title('pred mask : Binarization')
plt.imshow(mask[sid, :, :], cmap='gray')
plt.axis('off')

plt.subplot(122)
plt.title('pred mask : Remove small objects')
plt.imshow(pp_mask[sid, :, :], cmap='gray')
plt.axis('off')
plt.show()
```

```python
# fg

# 2020 origin, bad!
# # opening
# pp_fg = morphology.binary_opening(pp_mask, selem=morphology.cube(5))
# pp_fg = pp_fg.astype(np.uint8)
# # dilation
# pp_fg = morphology.binary_dilation(pp_fg, selem=morphology.ball(2))
# pp_fg = pp_fg.astype(np.uint8)

# new
# erosion
pp_fg = morphology.binary_erosion(pp_mask, morphology.cube(4))
pp_fg = pp_fg.astype(np.uint8)

# dilation
pp_fg = morphology.binary_dilation(pp_fg, morphology.cube(3))
pp_fg = pp_fg.astype(np.uint8)

plt.imshow(pp_fg[sid, :, :], cmap='gray')
plt.axis('off')
plt.show()
```

```python
# pp_bg

# dilation
pp_bg = morphology.binary_dilation(pp_mask, morphology.ball(3))

# bg dist_transform
dist_transform = distance_transform_edt(pp_bg)

# get bg watershed
labels = watershed(dist_transform)
boundary_image = find_boundaries(pp_bg, labels.all())
boundary_image = boundary_image.astype(np.uint8)

plt.subplot(121)
plt.imshow(dist_transform[sid, :, :])
plt.axis('off')
plt.subplot(122)
plt.imshow(boundary_image[sid, :, :])
plt.axis('off')
plt.show()
```

```python
pp_bg = pp_bg.astype(np.uint8)
unknow = cv.subtract(pp_bg, boundary_image)
unknow = cv.subtract(unknow, pp_fg)
unknow = unknow.astype(np.uint8)
print(np.unique(unknow))

plt.imshow(unknow[sid, :, :], cmap='gray')
plt.axis('off')
plt.show()
```

```python
# mwt input image
gradient = filters.sobel(pred_mask)

edge = pred_edge.copy()
mwt_input = gradient + edge

plt.imshow(mwt_input[sid, :, :], cmap='gray')
plt.axis('off')
```

```python
# Marker labelling
markers, num_markers = label(pp_fg)
markers = markers + 1
markers[unknow==1] = 0

plt.imshow(markers[sid, :, :])
plt.axis('off')
plt.show()
```

```python
labels = watershed(mwt_input, markers=markers, mask=pred_mask)
labels[labels == 1] = 0

print('gt_num_markers: ', gt_labels_num)
print('labels: ', len(np.unique(labels)))
plt.imshow(labels[sid, :, :])
plt.axis('off')
plt.show()
```

```python
# save labels
test_path = os.path.join('ToothDataset/cbcts')
out = sitk.GetImageFromArray(labels)

out.SetOrigin(gt_nii.GetOrigin())
out.SetDirection(gt_nii.GetDirection())
out.SetSpacing(gt_nii.GetSpacing())

out_name = '{}/{}_labels.nii.gz'.format(save_path, pid)

sitk.WriteImage(out, out_name)
print('save ', out_name)
```
